```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df['species'].unique()
```

```python
df.isnull().sum()
```

binary classifier korbo tai setosa hotaye dibo

```python
df = df[df['species']!='setosa']
```

```python
df['species']=df['species'].map({'versicolor':0, 'virginica':1})
```

```python
df.head()
```

split datasets into dependent and indepedent features

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=43)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'penalty':['l1','l2','elasticnet'], 'C':[1,2,3,4,5,6,8,10,20,30,40,50,60], 'max_iter':[100,200,300]}
```

```python
classifier_regressior = GridSearchCV(lr,param_grid=parameters, scoring='accuracy', cv=5)
```

```python
classifier_regressior.fit(x_train, y_train)
```

```python
print(classifier_regressior.best_params_)
```

```python
classifier_regressior.best_score_
```

```python
y_pred = classifier_regressior.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
score
```

```python
print(classification_report(y_pred, y_test))
```
